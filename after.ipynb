```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import ElasticNet, LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# ============================================
# 1. LOAD AND SPLIT DATA
# ============================================

# Load data
merged_data = pd.read_csv("C:\\Users\\vinay\\OneDrive\\Desktop\\hackathon\\autopharma\\dataset\\merged_data.csv")

# Define target and features
target = merged_data['LN_IC50']
features = merged_data.drop(columns=['LN_IC50'])
stratify_col = merged_data['TCGA_DESC']

print("Original data shape:", merged_data.shape)
print("Features shape:", features.shape)
print("Target shape:", target.shape)
print("\nTCGA_DESC distribution:")
print(stratify_col.value_counts())

# Split into training (60%) and temporary (40%) sets
split1 = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=42)

for train_index, temp_index in split1.split(features, stratify_col):
    X_train, X_temp = features.iloc[train_index], features.iloc[temp_index]
    y_train, y_temp = target.iloc[train_index], target.iloc[temp_index]
    stratify_temp = stratify_col.iloc[temp_index]

# Split temporary into validation (20%) and test (20%)
split2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)

for val_index, test_index in split2.split(X_temp, stratify_temp):
    X_val, X_test = X_temp.iloc[val_index], X_temp.iloc[test_index]
    y_val, y_test = y_temp.iloc[val_index], y_temp.iloc[test_index]

print("\n" + "="*50)
print("DATA SPLIT SUMMARY")
print("="*50)
print(f"Train: {X_train.shape[0]} samples ({X_train.shape[0]/len(features)*100:.1f}%)")
print(f"Val:   {X_val.shape[0]} samples ({X_val.shape[0]/len(features)*100:.1f}%)")
print(f"Test:  {X_test.shape[0]} samples ({X_test.shape[0]/len(features)*100:.1f}%)")

# ============================================
# 2. PREPROCESSING
# ============================================

# Identify numeric columns (exclude categorical like TCGA_DESC, DRUG_NAME, etc.)
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()
numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()

print(f"\nNumeric features: {len(numeric_cols)}")
print(f"Categorical features: {len(categorical_cols)}")

# Handle categorical columns (one-hot encoding or dropping)
if categorical_cols:
    print(f"\nCategorical columns found: {categorical_cols[:5]}...")
    # Option 1: Drop categorical columns for simple model
    X_train_num = X_train[numeric_cols]
    X_val_num = X_val[numeric_cols]
    X_test_num = X_test[numeric_cols]
    
    # Option 2: One-hot encode (uncomment if you want to use categorical features)
    # X_train_num = pd.get_dummies(X_train, columns=categorical_cols, drop_first=True)
    # X_val_num = pd.get_dummies(X_val, columns=categorical_cols, drop_first=True)
    # X_test_num = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True)
    # # Align columns
    # X_val_num = X_val_num.reindex(columns=X_train_num.columns, fill_value=0)
    # X_test_num = X_test_num.reindex(columns=X_train_num.columns, fill_value=0)
else:
    X_train_num = X_train
    X_val_num = X_val
    X_test_num = X_test

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_num)
X_val_scaled = scaler.transform(X_val_num)
X_test_scaled = scaler.transform(X_test_num)

print(f"\nFinal feature dimensions after preprocessing: {X_train_scaled.shape[1]}")

# ============================================
# 3. MODEL TRAINING & EVALUATION
# ============================================

from scipy.stats import pearsonr

def evaluate_model(model, X, y, set_name=""):
    """Evaluate model and return metrics"""
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    pearson_corr, pearson_pval = pearsonr(y, y_pred)
    
    print(f"\n{set_name} Metrics:")
    print(f"  RMSE:    {rmse:.4f}")
    print(f"  MAE:     {mae:.4f}")
    print(f"  R²:      {r2:.4f}")
    print(f"  Pearson: {pearson_corr:.4f} (p={pearson_pval:.2e})")
    
    return {'RMSE': rmse, 'MAE': mae, 'R2': r2, 'Pearson': pearson_corr, 'predictions': y_pred}

print("\n" + "="*50)
print("MODEL TRAINING")
print("="*50)

results = {}

# ============================================
# BASELINE: Linear Regression
# ============================================
print("\n[1/4] Training Linear Regression (Baseline)...")
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)

results['Linear Regression'] = {
    'model': lr_model,
    'train': evaluate_model(lr_model, X_train_scaled, y_train, "Train"),
    'val': evaluate_model(lr_model, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(lr_model, X_test_scaled, y_test, "Test")
}

# ============================================
# ElasticNet
# ============================================
print("\n[2/4] Training ElasticNet...")
elastic_model = ElasticNet(alpha=0.1, l1_ratio=0.5, max_iter=10000, random_state=42)
elastic_model.fit(X_train_scaled, y_train)

results['ElasticNet'] = {
    'model': elastic_model,
    'train': evaluate_model(elastic_model, X_train_scaled, y_train, "Train"),
    'val': evaluate_model(elastic_model, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(elastic_model, X_test_scaled, y_test, "Test")
}

# ============================================
# Random Forest
# ============================================
print("\n[3/4] Training Random Forest...")
rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=15,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train_scaled, y_train)

results['Random Forest'] = {
    'model': rf_model,
    'train': evaluate_model(rf_model, X_train_scaled, y_train, "Train"),
    'val': evaluate_model(rf_model, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(rf_model, X_test_scaled, y_test, "Test")
}

# ============================================
# XGBoost
# ============================================
print("\n[4/4] Training XGBoost...")
xgb_model = XGBRegressor(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
xgb_model.fit(X_train_scaled, y_train)

results['XGBoost'] = {
    'model': xgb_model,
    'train': evaluate_model(xgb_model, X_train_scaled, y_train, "Train"),
    'val': evaluate_model(xgb_model, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(xgb_model, X_test_scaled, y_test, "Test")
}

# ============================================
# HYPERPARAMETER TUNING (using validation set)
# ============================================
print("\n" + "="*50)
print("HYPERPARAMETER TUNING")
print("="*50)

from sklearn.model_selection import RandomizedSearchCV

# Tune Random Forest
print("\nTuning Random Forest...")
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [5, 10, 20],
    'min_samples_leaf': [2, 5, 10]
}

# Use validation set for tuning by creating custom CV split
from sklearn.model_selection import PredefinedSplit
# Combine train and val for tuning
X_train_val = np.vstack([X_train_scaled, X_val_scaled])
y_train_val = np.concatenate([y_train, y_val])
# Create split indices: -1 for train, 0 for val
split_index = [-1] * len(X_train_scaled) + [0] * len(X_val_scaled)
ps = PredefinedSplit(test_fold=split_index)

rf_random = RandomizedSearchCV(
    RandomForestRegressor(random_state=42, n_jobs=-1),
    param_distributions=rf_param_grid,
    n_iter=10,
    cv=ps,
    scoring='neg_root_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
rf_random.fit(X_train_val, y_train_val)

print(f"Best RF params: {rf_random.best_params_}")
print(f"Best RF validation RMSE: {-rf_random.best_score_:.4f}")

results['Random Forest (Tuned)'] = {
    'model': rf_random.best_estimator_,
    'val': evaluate_model(rf_random.best_estimator_, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(rf_random.best_estimator_, X_test_scaled, y_test, "Test")
}

# Tune XGBoost
print("\nTuning XGBoost...")
xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

xgb_random = RandomizedSearchCV(
    XGBRegressor(random_state=42, n_jobs=-1),
    param_distributions=xgb_param_grid,
    n_iter=10,
    cv=ps,
    scoring='neg_root_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
xgb_random.fit(X_train_val, y_train_val)

print(f"Best XGB params: {xgb_random.best_params_}")
print(f"Best XGB validation RMSE: {-xgb_random.best_score_:.4f}")

results['XGBoost (Tuned)'] = {
    'model': xgb_random.best_estimator_,
    'val': evaluate_model(xgb_random.best_estimator_, X_val_scaled, y_val, "Validation"),
    'test': evaluate_model(xgb_random.best_estimator_, X_test_scaled, y_test, "Test")
}

# ============================================
# 4. RESULTS SUMMARY
# ============================================
print("\n" + "="*50)
print("FINAL RESULTS SUMMARY")
print("="*50)

summary_df = pd.DataFrame({
    'Model': [],
    'Val_RMSE': [],
    'Val_Pearson': [],
    'Val_R2': [],
    'Test_RMSE': [],
    'Test_Pearson': [],
    'Test_R2': []
})

for model_name, result in results.items():
    summary_df = pd.concat([summary_df, pd.DataFrame({
        'Model': [model_name],
        'Val_RMSE': [result['val']['RMSE']],
        'Val_Pearson': [result['val']['Pearson']],
        'Val_R2': [result['val']['R2']],
        'Test_RMSE': [result['test']['RMSE']],
        'Test_Pearson': [result['test']['Pearson']],
        'Test_R2': [result['test']['R2']]
    })], ignore_index=True)

print("\n", summary_df.to_string(index=False))

# Find best model based on validation RMSE
best_model_name = summary_df.loc[summary_df['Val_RMSE'].idxmin(), 'Model']
print(f"\n🏆 Best Model (by Validation RMSE): {best_model_name}")
print(f"   Test RMSE:    {results[best_model_name]['test']['RMSE']:.4f}")
print(f"   Test Pearson: {results[best_model_name]['test']['Pearson']:.4f}")
print(f"   Test R²:      {results[best_model_name]['test']['R2']:.4f}")

# ============================================
# 5. FEATURE IMPORTANCE (for tree-based models)
# ============================================
if best_model_name in ['Random Forest', 'XGBoost']:
    print("\n" + "="*50)
    print("TOP 10 FEATURE IMPORTANCES")
    print("="*50)
    
    importances = results[best_model_name]['model'].feature_importances_
    feature_importance_df = pd.DataFrame({
        'Feature': numeric_cols,
        'Importance': importances
    }).sort_values('Importance', ascending=False).head(10)
    
    print("\n", feature_importance_df.to_string(index=False))

print("\n" + "="*50)
print("PIPELINE COMPLETE!")
print("="*50)
```
